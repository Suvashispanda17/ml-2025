```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import load_digits
```

```python
dig = load_digits()
dir(dig)
```

```python
%matplotlib inline 
plt.gray()
for i in range(4):
    plt.matshow(dig.images[i])
```

```python
df = pd.DataFrame(dig.data)
df.head()
```

```python
df["target"] = dig.target
df.head()
```

```python
x = df.drop(["target"],axis ="columns")
x
```

```python
y=  df.target
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train ,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.1)
```

```python
len(x_test)
```

```python
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier(n_estimators=40)   #n_estimators=20 is used for tuning means takes 20 decision trees for making best prediction
```

```python
model.fit(x_train,y_train)
```

```python
y_pred = model.predict(x_test)
y_pred
```

```python
model.score(x_test,y_test)*100
```

```python
from sklearn.metrics import confusion_matrix
cn =confusion_matrix(y_test,y_pred)
cn
```

```python
import seaborn as sns
```

```python
plt.figure(figsize =(10,7))
sns.heatmap(cn , annot =True)
```
