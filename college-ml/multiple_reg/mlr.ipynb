```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
data = pd.DataFrame(dict(area =[2600,3600,4600,5600,6600,7600],
                        price =[250000,350000,450000,550000,650000,750000],
                        bed_room=[6,8,5,7,2,4],
                        age=[20,21,22,23,25,24,]))
```

```python
data
```

```python
# data.to_csv("multiple.csv",index=False)
```

```python
df = pd.read_csv("multiple.csv")
df
```

```python
import math as mt 
med = mt.floor(df.bed_room.median())
med
```

```python
df.bed_room =df.bed_room.fillna(med)
df
```

```python
reg = linear_model.LinearRegression()
```

```python
reg.fit(df[["area","bed_room","age"]],df.price)
```

```python
reg.predict([[6000,5,34]])
```

```python
reg.coef_ , reg.intercept_
```

```python
import pickle
```

```python
with open('model_pickle','wb') as f:  #wb if write binary 
    pickle.dump(reg,f)
    
```

```python
with open('model_pickle','rb') as f : # rb is  read binary
    model = pickle.load(f)
    
```

```python
model.predict([[6000,6,6]])
```
