```python
import pandas as pd
import matplotlib.pyplot as plt 
```

```python
data = pd.read_csv("insurance_data.csv")
data.head(6)
```

```python
plt.scatter(data.age ,data.bought_insurance,marker =">" ,color ="green")
```

```python
from sklearn import model_selection
```

```python
x_train ,x_test , y_train, y_test = model_selection.train_test_split(data[["age"]],data.bought_insurance,test_size = 0.2)
```

```python
#import logistic regression
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
reg = LogisticRegression()
reg
```

```python
reg.fit(x_train,y_train)
```

```python
reg.predict(x_test)
```

```python
reg.predict([[57]])
```

```python
import numpy as np
plt.scatter(data.age ,data.bought_insurance,marker =">" ,color ="green")
plt.plot(data.age ,reg.predict(np.array(data.age).reshape(-1, 1)) ,color ="green")
```

```python
reg.score(x_test,y_test)*100
```
