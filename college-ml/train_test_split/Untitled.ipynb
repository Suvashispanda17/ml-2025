```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv("car.csv")
data.head(5)
```

```python
from matplotlib import pyplot as plt
```

```python
%matplotlib inline
```

```python
plt.scatter(data.Mileage ,data["Sell Price($)"],marker =">",color ="blue")
```

```python
plt.scatter(data["Age(yrs)"] ,data["Sell Price($)"],marker =">",color ="blue")
```

```python
x  = data[["Mileage",	"Age(yrs)"]]
y = data[["Sell Price($)"]]
```

```python
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train ,x_test,y_train, y_test=  train_test_split(x,y,test_size=0.2)   #test_size = 0.2  means 0.2 * 100  == 20%  demonstrating 80 % is training data and 20% is testing data
```

```python
print(f"length  of x train is : {len(x_train)}   , length  of x test is : {len(x_test)}, length  of y train is : {len(y_train)} , length  of y test is : {len(y_test)}")
```

```python
from sklearn import linear_model
```

```python
reg = linear_model.LinearRegression()
```

```python
reg.fit(x_train,y_train)
reg
```

```python
reg.predict(x_test)
```

```python
y_test
```

```python
print(f"The Accuracy is: {reg.score(x_test,y_test)*100}")
```
