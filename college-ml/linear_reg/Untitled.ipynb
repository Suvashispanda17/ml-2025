```python
import pandas as pd 
from matplotlib import pyplot as plt 
from sklearn import linear_model
```

```python
data = pd.read_csv("linear.csv")
data
```

```python
%matplotlib inline
plt.xlabel("area")
plt.ylabel("price")
plt.scatter(data.Area,data.price , color ="green" , marker ="o")
```

```python
reg = linear_model.LinearRegression()
reg.fit(data[["Area"]], data.price)
```

```python
reg.predict([[650]])
```

```python
reg.coef_  #this is m 
```

```python
reg.intercept_  #this is slope
```
