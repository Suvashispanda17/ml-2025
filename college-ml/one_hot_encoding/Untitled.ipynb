```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv("homeprices.csv")
data

```

```python
dummy = pd.get_dummies(data.town,dtype =int)
dummy
```

```python
merge = pd.concat([data,dummy],axis ="columns")
merge
```

```python
final = merge.drop(["town","west windsor"],axis ="columns")
```

```python
final
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
```

```python
x = final.drop(['price'],axis ="columns")
x
```

```python
y = final.price
y
```

```python
model.fit(x,y)
```

```python
model.predict([[3600,1,0]])
```

```python
model.score(x,y)*100
```
